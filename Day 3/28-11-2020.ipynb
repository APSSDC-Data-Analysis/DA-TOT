## Data Cleaning Using Pandas
### Working on Duplicate Values and Handling Missing Values
- There are mainly 4 methods to identify & manipulating missing values
    - isnull(): Presence of missing values
    - notnull(): opposite to isnull()
    - dropna(): It deletes null values
    - fillna(): Replace the missing values
    
- ***Note: Always have backup of original dataset***
- ***Warning: Do not run a program more than one time***

```python
dict1 = {'Trainers':['Niharika','Srilalitha','Satheesh','Vijay','Vijay'], 
         'Subjects':['ML','DS','AI','DA','DA']}
data = pd.DataFrame(dict1)
```

```python
data
```

```python
## Displays duplicate values if data set contains
data[data.duplicated()]
```

```python
data.drop_duplicates(inplace = True)
```

```python
data
```

### Replace missing values
- Replace missing values either with "0" or mean() value, if that data is numerical
- Replace missing values with "most occured object value" or any object like " No_name, None..." 
    - fillna()

```python
data2 = market_df.copy()
```

```python
data2.head()
```

```python
data2.shape
```

```python
data2.isnull().sum()
```

```python
data2['Product_Base_Margin'].mean()
```

```python
data2['Product_Base_Margin'] = data2['Product_Base_Margin'].fillna(data2['Product_Base_Margin'].mean())
```

```python
data2.shape
```

```python
data2.isnull().sum()
```
