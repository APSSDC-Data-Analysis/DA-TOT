````python
input_featues = cancer[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'radius_se', 'perimeter_se',
       'area_se', 'concave points_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
````

````python
output_features = cancer['diagnosis']
````

````python
# Seperate the data into train and test data
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te=train_test_split(input_featues,output_features,
                                    test_size=0.3,
                                    random_state=2)
````

````python
# importing required Algorithm
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
````

````python
# train the model using fit method
log.fit(x_tr,y_tr)
````

````python
# predict the values for testing purpose
pred_values = log.predict(x_te)
````

````python
# Evaluate the model
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_te,pred_values)
````

Change the random_state and test_size paramer values check the accuracy

````python
confusion_matrix(y_te,pred_values)
````

````python
tt.drop('cabin',inplace=True,axis=1)
````

````python
tt.drop('name',axis=1,inplace=True)
````

````python
tt.columns
````

````python
from sklearn.preprocessing import LabelEncoder
lae = LabelEncoder()
````

````python
tt['sex']=lae.fit_transform(tt['sex'])
tt['sex']
````

````python
tt['age']
````

````python
tt['age']=tt['age'].fillna(round(tt['age'].mean()))
````

````python
round(tt.age.mean())
````

````python
tt.age.isna().sum()
````

````python
tt['ticket'] = lae.fit_transform(tt['ticket'])
tt['ticket']
````

````python
tt.embarked
````

````python
tt.embarked.isna().sum()
````

````python
tt.embarked.unique()
````

````python
tt.embarked.value_counts()
````

````python
tt['embarked'] = tt.embarked.fillna('S')
tt['embarked'].isna().sum()
````

````python
tt['embarked'] = lae.fit_transform(tt['embarked'])
tt['embarked']
````

````python
tt.info()
````

````python
# seperate the data into input and output labels
input_labels = tt.drop('survived',axis=1)
input_labels.head()
````

````python
output_labels = tt['survived']
````

````python
# seperate data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
input_labels,output_labels,random_state=10,test_size=0.3)
````

````python
# import required Algorithm
from sklearn.svm import SVC
sv = SVC(kernel='linear',random_state=0)
````

````python
# train the model
sv.fit(x_train,y_train)
````

````python
# predict the values for testing
pred = sv.predict(x_test)
````

````python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,pred)
````

````python
confusion_matrix(y_test,pred)
````
