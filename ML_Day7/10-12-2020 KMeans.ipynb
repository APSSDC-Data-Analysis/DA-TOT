````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
mall_data = pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Clustering/Mall_Customers.csv")
mall_data.head()
````

````python
mall_data.columns
````

````python
mall_data_dum = mall_data.drop('CustomerID',axis=1)
mall_data_dum.head()
````

````python
mall_data_dum.isna().sum()
````

````python
mall_data_dum.info()
````

````python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
````

````python
mall_data_dum['Genre'] = lbe.fit_transform(
    mall_data_dum['Genre'])
````

````python
mall_data_dum['Genre'].unique()
````

````python
mall_data_dum['Genre'].value_counts()
````

````python
mall_data_dum
````

Check types of scallers and know about each scaller

````python
from sklearn.model_selection import train_test_split
data_train,data_test = train_test_split(mall_data_dum,
                                        random_state=2,
                                        test_size=0.25)
````

````python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
````

````python
km.fit(data_train)
````

````python
help(km.fit(data_train))
````

````python
pred = km.predict(data_train)
pred
````

````python
pred_test = km.predict(data_test)
pred_test
````
