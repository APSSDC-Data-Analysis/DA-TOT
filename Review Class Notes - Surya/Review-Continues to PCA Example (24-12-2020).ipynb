```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Get the data:
dataset=pd.read_csv("wine.csv")
```

```python
#Proprecessing:
dataset.isna().sum()
```

```python
#Separate input features and output features:
X = dataset.iloc[:,:-1].values #input features
y = dataset.iloc[:,-1].values #target features
```

```python
#Separate dataset into train data and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Feature scaling:
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#Load the PCA Model:
from sklearn.decomposition import PCA
pca=PCA(n_components = 2)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
variance = pca.explained_variance_ratio_
```

```python
variance
```

```python
#We have to check our PCA is working or not by taking a new classier model.
```

```python
from sklearn.linear_model import LogisticRegression
cl = LogisticRegression()
cl.fit(x_train,y_train)
```

```python
y_pred = cl.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```

```python
accuracy_score(y_test,y_pred)
```
