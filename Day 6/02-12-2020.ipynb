### Pie Chart
- plt.pie()

```python
sports = ['cricket','chess','football','ludo']
score = [80,90,50,75]
plt.pie(score, labels = sports, autopct = '%1.1f%%', 
        colors = ['c','m','pink','b'], startangle = 90, explode = (0,0,0,0.2),
       shadow = True)
plt.show()
```

```python
help(plt.pie)
```

### Area Plot
- plt.stackplot()

```python
mrng_shift = [1,2,3,4]
eng_shift = [1,4,3,1]
night_shift = [3,1,5,1]
plt.stackplot(mrng_shift,eng_shift, night_shift, colors = ['c','r'])
```

## Categorical Plot
- By default it returns a scatter plot
    - sns.catplot()

```python
help(sns.catplot)
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = iris)
```

```python
## It display only for numerical columns
sns.catplot(data = iris)
```

```python
sns.catplot(x = 'species', y = 'sepal_length', 
            data = iris, palette = 'Blues')
```

- Boxplot

```python
sns.catplot(data = iris, kind = 'box', orient = 'h', palette = 'husl')
```

- Violin Plot
    - similar to boxplot

```python
sns.catplot(x = 'species',y = 'sepal_width',
            data = iris, kind = 'violin', palette = 'husl')
```

- strip plot
    - It is represented the data in sorted oreder

```python
sns.catplot(x = 'species',y = 'petal_length', 
            data = iris, kind = 'strip')
```

- swarm plot
    - Also known as 'bees plot'
    - Avoids Overlapping

```python
sns.catplot(x = 'species',y = 'petal_length',data =iris, kind = 'swarm')
```

```python
sns.catplot(x = 'species',y = 'petal_width',data =iris, kind = 'swarm')
```

### Joint Plot
- Combination of 2 plots, by default it contains scatter and histogram plots
    - sns.jointplot()

```python
iris.head(2)
```

```python
sns.jointplot(x = 'sepal_length',y = 'sepal_width',data = iris)
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = 'sepal_length',y = 'sepal_width',
              data = iris, kind = 'hex', color = 'orange')
```

```python
sns.jointplot(x = 'sepal_length',y = 'sepal_width',
              data = iris, kind = 'kde', color = 'm')
```

### Pair Plot
- It builds a histogram and scatter plot for pair wise relationship
- sns.pairplot()

```python
help(sns.pairplot)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
sns.pairplot(iris, vars = ['sepal_length','sepal_width'])
```

```python
sns.pairplot(iris, hue = 'species', palette = 'husl')
```

```python
sns.pairplot(iris, hue="species", markers=["h", "+", "D"])
```

### Heatmaps
- Heatmap displays matrix-like graph, where cells are filled with colors based on values.
- sns.heatmap()

```python
t = sns.load_dataset('tips')
```

```python
t.head()
```

```python
t.iloc[:,[0,1,-1]].head()
```

```python
help(sns.heatmap)
```

```python
sns.heatmap(t.iloc[:,[0,1,-1]].head(), annot = True)
```

```python
t.corr()
```

```python
sns.heatmap(t.corr(),annot = True, cmap = 'Blues')
```
