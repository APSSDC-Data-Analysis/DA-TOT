### combining and merging of dataframes:


```python
a = {"name":["niharikaa","mounika","lavanya"],
    "Branch":["Ece","cse","ece"],
    "Gender":["F","F","F"]}
a = pd.DataFrame(a,index = [121,122,123])
```

```python
a
```

```python
pd.concat([a,a])  ### rows concate
```

```python
pd.concat([a,a],axis = 1)  ### column concate
```

```python
a.append(a)  ### only wise appending
```

```python
pd.merge(a,a)
```

```python
d1 = pd.DataFrame({"std_name":['r','b','v','t'],
                  "age":[6,7,8,2]})
d1
```

```python
d2 = pd.DataFrame({"std_name":['r','a','b','z'],
                  "fav_fruit":["mango","banana","apple","grapes"]})
d2
```

```python
pd.concat([d1,d2])
```

```python
pd.merge(d1,d2)   ### common columns mmerging
```

```python
pd.merge(d1,d2,how = "left")
```

```python
pd.merge(d1,d2,how = "right")
```

```python
pd.merge(d1,d2,how = "outer")   ### union of both dataframes
```

```python
pd.merge(d1,d2,how = "inner")   ## intersection of both dataframes
```

```python
help(pd.merge)
```
